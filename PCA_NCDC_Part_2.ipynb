```python
#Provided by prof freund
def loads(eVal):
    return pickle.loads(zlib.decompress(base64.b64decode(eVal)))
def dumps(Value):
    return base64.b64encode(zlib.compress(pickle.dumps(Value),9))
```

##Loading Covariance components
First we must load the covariance matricies from our file.



```python
file_path = "mr_output/make_covar.raw_256.bak"
```

```python
f = open(file_path,'r')
line = f.next()
while(line[0:4] != '"ok"'):
    line = f.next()
f.close()
pat = re.compile('"ok"\t\[(\d*), "(.*)", "(.*)"\]$')
mat = pat.match(line.strip())
n, vec, outer = mat.groups()
n     = int(n)
vec   = loads(vec)
outer = loads(outer)
(n,vec.shape,vec[0:10],outer.shape,outer[0:10,0])
```

```python
covar = (outer - np.outer(vec,vec))/n
```

```python
U,D,V=np.linalg.svd(covar)
```

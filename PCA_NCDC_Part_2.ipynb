##Loading Covariance components
First we must load the covariance matricies from our file.



```python
file_path = "mr_output/make_covar.raw_256.bak"
```

```python
f = open(file_path,'r')
line = f.next()
while(line[0:4] != '"ok"'):
    line = f.next()
f.close()
pat = re.compile('"ok"\t\[(\d*), "(.*)", "(.*)"\]$')
mat = pat.match(line.strip())
n, vec, outer = mat.groups()
n     = int(n)
vec   = loads(vec)
outer = loads(outer)
(n,vec.shape,vec[0:10],outer.shape,outer[0:10,0])
```

Here we compute our covariance matrix and then the eigen vector (U), relative error (D), and eigen values for each data entry (V)

```python
covar = (outer/n - np.outer(vec/n,vec/n))
```

```python
U,D,V=np.linalg.svd(covar)
```

```python
D[0:10]
```

```python
err = np.cumsum(D[:])/np.sum(D[:])
plt.plot(err)
k = [ n for n,i in enumerate(err) if i>.95 ][0]
print(str(k) + " eigen vectors needed for 95% accuracy")
#plt.xlim((-1, 20))
#plt.ylim((0, 1.1))
```

Here we plotted the error imrpvoement for all eigen values, with just the mean we get a ~75% reconstruction error and we need 24 eigen vectors to reach 95% accuracy.  This seems reasonable for a _global_ model (without region partitioning).
Now we plot the mean and then first 10 eigen values for analysis:

```python
plt.plot(vec.transpose()/n)
plt.title("Mean")
for i in range(10):
    plt.figure()
    plt.plot(U[:,i])
    plt.title("Eigen value " + str(i))
```

It looks like after the third eigen vector noise begins to dominate on temperature, the remaining ones seem to fit the snowdepth variable.

This makes sense since we did not z-scale vectors, so the temperature values are much larger range than the snowdepth/percipitation so fixing them is the first priority in the PCA.

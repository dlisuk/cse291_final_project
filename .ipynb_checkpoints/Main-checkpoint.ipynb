Now lets have a look at the actual measurements.

Below we have picked the following 4 types of measurements:

```python
head = ["t_mean","dewp","slp","stp","visib","ws_mean","ws_max",
        "ws_gust", "t_max", "t_min", "prcp","snwd", "weather_type"]

c = {}
f = open("mr_output/ncdc_count","r")
for line in f:
    if( "error" in line):
        continue
    F = line.translate(string.maketrans(",","\t"),'[]" \n').split('\t')
    if(F[0] == "NaN" or F[1] == "NaN"):
        continue
    x = float(F[0])
    t = int(x/10)*10+cmp(x,0)*5
    x = float(F[1])
    n = int(x/10)*10+cmp(x,0)*5
    c[(t,n)] = c.get((t,n),[0 for x in head])
    c[(t,n)] = [x+y for x,y in zip(c[(t,n)],[float(x) for x in F[2:]])]
xp=[]
yp=[]
zp=[]
for ((t,n),v) in c.items():
    xp.append(n)
    yp.append(t)
    zp.append(v)
```

```python
for i in [0,5,8,10]: # t_mean, ws_mean, t_max, prcp
    plt.figure(figsize=(15,10),dpi=300)
    m = Basemap(projection='merc',llcrnrlat=latsmin,urcrnrlat=latsmax,\
                llcrnrlon=lonmin,urcrnrlon=lonmax,lat_ts=20,resolution='i')

    m.drawcoastlines()
    m.fillcontinents(color='coral',lake_color='aqua')

    # draw parallels and meridians.
    parallels = np.arange(-80,81,10.)
    # labels = [left,right,top,bottom]
    m.drawparallels(parallels,labels=[False,True,True,False])
    meridians = np.arange(10.,351.,10.)
    m.drawmeridians(meridians,labels=[True,False,False,True])

    #m.drawparallels(np.arange(-90.,91.,30.))
    #m.drawmeridians(np.arange(-180.,181.,60.))
    m.drawmapboundary(fill_color='aqua')

    # draw map with markers for locations
    x, y = m(xp,yp)
    c = np.log10([z[i]/365 for z in zp])
    cs = m.scatter(x,y,c=c,marker='s', lw = 0,zorder=2, s=40)
    cbar = m.colorbar(cs,location='bottom',pad="5%")
    cbar.set_label("Data Years (log scale)")
    plt.title(head[i])
    plt.show()
```

In general this shows that the US and Europe are the places with most measurements recorded.

##PCA analysis on areas with similar weather pattern

In this section we perform a PCA analysis on the data. In order to get a good fit that works for the entire world, we partition the data into smaller areas where the data might have similar patterns and perform a PCA on each area. We then combine areas that have similar weather patterns so that we get a more compact representation of the data.

This section consists of three parts:
1. Hierarchically splitting the data
2. Performing PCA on each area
3. Merging the areas based on the Minimum Decsribtion Length principle

###Hierarchically splitting the data
_In this notebook we uses the splits that was created in [THIS NOTEBOOK](Splitting%20the%20stations.ipynb) ([nbviewer](http://nbviewer.ipython.org/github/dlisuk/cse291_final_project/blob/master/Splitting%20the%20stations.ipynb)). The full description of the process of splitting of both data set can be seen in that notebook._

If we blindly split the world into small tiles, we might end up having areas with few (or no) data points and areas with many data point, leading to overfitting in some parts of the world and not being able to model local weather phenomenons in other parts of the world. We would rater partition the data into areas of equal number of measurements.

We can do a hierarchical splitting of geospatial data by dividing the data points recursively in a R-tree ([Wikipedia](http://en.wikipedia.org/wiki/R-tree)). There are different heuristics defining how to split the data. Here we will just take a simple approach and split the data recursively into two parts with alternating latitude and longitude splits.

Lets load in the data with hierarchical labels that we created:

```python
ncdc = pickle.load(open('local_data/ncdc_data_labels.pkl', 'r'))
ncdc.head()
```

We are also going to need a function to cut the R-tree in any level and get a unique label for each area

```python
def unique_label(row, level):
    return hash(row[['level_%d'%l for l in range(level+2)]].astype(str).sum())
def unique_labels(dataframe, level):
    return dataframe.apply(lambda x: unique_label(x, level=level), axis=1)
```

...and the function to create the basemap plots.

```python
def color_map(dataframe, lon_col, lat_col, color, alpha=1):
    plt.figure(figsize=(15,10),dpi=300)
    lon = dataframe[lon_col]
    lat = dataframe[lat_col]
    lonmin=-180
    lonmax=180
    latsmin=-80
    latsmax=80
    
    m = Basemap(projection='merc',llcrnrlat=latsmin,urcrnrlat=latsmax,\
        llcrnrlon=lonmin,urcrnrlon=lonmax,lat_ts=20,resolution='c')
    #m.drawcoastlines(color='lightgrey')
    m.drawmapboundary()
    
    x,y = m(np.array(lon), np.array(lat))
    color_map = m.scatter(x,y,c=np.array(color),marker='.', lw=0, alpha=alpha)
    m.colorbar(color_map)
```

Here is the 4th level split:

```python
color_map(ncdc, 'longitude', 'latitude', unique_labels(ncdc, level=4))
```

And this is what the 10th level looks like:

```python
color_map(ncdc, 'longitude', 'latitude', unique_labels(ncdc, level=10))
```

- SMOOTHING and efficiency
PCA_NCDC.ipynb

To compute PCA you must take the eigen values of the covariance matrix.
In a streaming setting, the covariance matrix is best computed via the following formula:

$$Cov(x) = \mathbf{E}(x\circ x) - \mathbf{E}(x)\circ \mathbf{E}(x)$$
... where $\circ$ denotes the outer product.

Effectively we need the mean outer product of all data vectors minus the outer product of the mean of the vectors. This should be done  seperately in each area so we need the following mapreduce job:

**Mapper:** Filter out the measurements we are interessted in and output a key-value pair with the area label as the key, and the measurements vector as the value.

**Reducer:** (Called for each area) Compute the covariance matrix

**Combiner:** This is not needed, but to improve efficiency we perform a reduce step.

```python
The singular value vector, D, contains th fraction of variance explained by each eigen vector (the columns of U).  To plot the percentage of variance explained by the first $x$ eigen vectors, we can plot $\frac{\sum_{i=0}^x D[i]}{\sum D}$.
```
